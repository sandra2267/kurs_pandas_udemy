```python
import pandas as pd
from datetime import datetime 
from datetime import timedelta 
import time
```

```python
df = pd.read_csv('/Users/sandramarczak/Documents/udemy/pandas_rafal_mobilo/course-files/marathon_results_2016.csv', index_col='Bib', usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'])
```

```python
df['40K'] = df['40K'].replace('-', 0)
df['Half'] = df['Half'].replace('-', 0)
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds (x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds (x))
```

```python
newDf = df.pivot_table(index='Age', columns='M/F', values = ['TotalSeconds', 'HalfSeconds'])
```

```python
newDf.swaplevel(axis='columns').sort_index(axis='columns')
```

```python
df.pivot_table(index="Age",columns="M/F",aggfunc='count',values='City')
```
