```python
import pandas as pd
import numpy as np
```

```python
fortune = pd.read_csv('../../dane/course-files/Fortune_500_2017.csv', usecols=['Rank','Title','Employees','Profits','Assets'], index_col='Rank')
```

```python
fortune.head()
```

```python
fortune['RankByEmployees'] = fortune['Employees'].rank(ascending=False)
```

```python
fortune.head()
```

```python
fortune['RankByProfits'] = fortune['Profits'].rank(ascending=False)
```

```python
fortune.head()
```

```python
isEmployeeRankFirst10 = fortune.RankByEmployees < 10
```

```python
isEmployeeRankFirst10.head()
```

```python
isProfitRankFirst10 = fortune.RankByProfits < 10
```

```python
isProfitRankFirst10.head()
```

```python
fortune[isEmployeeRankFirst10 & isProfitRankFirst10]
```

```python
isEmplolyeesRankMoreThan400 = fortune.RankByEmployees > 400
```

```python
fortune[isEmplolyeesRankMoreThan400]
```

```python
fortune[isEmplolyeesRankMoreThan400 & isProfitRankFirst10]
```
