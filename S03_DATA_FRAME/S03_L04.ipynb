```python
fuel = pd.read_csv('fuel.csv', low_memory=False, usecols=['Vehicle ID','Year','Make','Model','Class','Fuel Type','Combined MPG (FT1)'])
fuel.head()
```

```python
fuel['CombinedKPL'] = fuel['Combined MPG (FT1)']*1.609/3.78
```

```python
fuel.insert(loc = 5, column = 'litersPerKm', value= 100/fuel['CombinedKPL'])
```
